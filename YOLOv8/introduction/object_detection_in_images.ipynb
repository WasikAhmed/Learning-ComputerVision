## Importing Modules

```python
from PIL import Image
from ultralytics import YOLO

YOLO
```

## Creating a neural network model

```python
model = YOLO('yolo-Weights/yolov8m.pt')
```

```python
model.names
```

## Detecting Objects

```python
results = model.predict("images/cat_dog.jpg")
```

```python
result = results[0]
result
```

```python
len(result.boxes)
```

```python
box = result.boxes[0]
box
```

```python
box.cls
```

```python
box.xyxy
```

```python
box.conf
```

```python
cords = box.xyxy[0].tolist()
class_id = box.cls[0].item()
conf = box.conf[0].item()

cords, class_id, conf
```

```python
result.names
```

```python
cords = box.xyxy[0].tolist()
cord = [round(x) for x in cords]
class_id = result.names[box.cls[0].item()]
conf = round(box.conf[0].item(), 2)
cords, class_id, conf
```
